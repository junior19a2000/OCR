```python
import ipywidgets as ipw
import bqplot as bqp
import openpyxl

datos     = openpyxl.load_workbook(r'D:\Trabajo\Base_de_datos.xlsx')
sheet_dat = datos.active
```

```python
def open_docx(docx):
    datos        = openpyxl.load_workbook(r'D:\Trabajo\Base_de_datos.xlsx')
    sheet_dat    = datos.active
    expedientes  = list(Dropdown1.options)
    expediente   = expedientes.index(docx)
    cell_obj     = sheet_dat.cell(row = int(expediente) + 2, column = 5)
    url          = cell_obj.value.replace(' ','')
    urls         = "https://drive.google.com/file/d" + url + "preview"
    HTML5.value  = f"<iframe src={urls} width={width} height={height}>"
    cell_obj     = sheet_dat.cell(row = int(expediente) + 2, column = 1)
    Text1.value  = str(cell_obj.value)
    cell_obj     = sheet_dat.cell(row = int(expediente) + 2, column = 2)
    Text2.value  = cell_obj.value
    cell_obj     = sheet_dat.cell(row = int(expediente) + 2, column = 3)
    Text3.value  = cell_obj.value
    
def filtrar(case):
    datos     = openpyxl.load_workbook(r'D:\Trabajo\Base_de_datos.xlsx')
    sheet_dat = datos.active
    valores   = [''] * (sheet_dat.max_row - 1)
    for i in range(2, sheet_dat.max_row + 1):
        cell_obj       = sheet_dat.cell(row = i, column = 2)
        mat            = cell_obj.value
        valores[i - 2] = str(mat)
    indices = []
    for valor in range(len(valores)):
        if valores[valor] == case:
            indices.append(valor)
    valores = [''] * (sheet_dat.max_row - 1)
    for i in range(2, sheet_dat.max_row + 1):
        cell_obj       = sheet_dat.cell(row = i, column = 1)
        exp            = cell_obj.value
        valores[i - 2] = str(exp)
    values  = [valores[i] for i in indices]
    Dropdown3.options = values
    
def add_data(click):
    datos     = openpyxl.load_workbook(r'D:\Trabajo\Base_de_datos.xlsx')
    sheet_dat = datos.active
    valores = [''] * (sheet_dat.max_row - 1)
    for i in range(2, sheet_dat.max_row + 1):
        cell_obj       = sheet_dat.cell(row = i, column = 1)
        exp            = cell_obj.value
        valores[i - 2] = str(exp)
    if Text1.value in valores:
        Text4.value = 'Expediente ' + Text1.value + ' ya existe'
    else:
        fila = sheet_dat.max_row + 1
        if Dropdown4.value == 'Caso desconocido':
            cell_obj        = sheet_dat.cell(row = fila, column = 1)
            cell_obj.value  = int(Text1.value)
            cell_obj        = sheet_dat.cell(row = fila, column = 2)
            cell_obj.value  = Text2.value
            cell_obj        = sheet_dat.cell(row = fila, column = 3)
            cell_obj.value  = Text3.value
            cell_obj        = sheet_dat.cell(row = fila, column = 5)
            cell_obj.value  = Text4.value
        elif Dropdown4.value == 'Caso conocido':
            cell_obj        = sheet_dat.cell(row = fila, column = 1)
            cell_obj.value  = int(Text1.value)
            cell_obj        = sheet_dat.cell(row = fila, column = 2)
            cell_obj.value  = Dropdown2.value
            cell_obj        = sheet_dat.cell(row = fila, column = 3)
            cell_obj.value  = Dropdown5.value
            cell_obj        = sheet_dat.cell(row = fila, column = 5)
            cell_obj.value  = Text4.value
        Text4.value = 'Expediente ' + Text1.value + ' añadido'
        datos.save(r'D:\Trabajo\Base_de_datos.xlsx')
        valores = list(Dropdown1.options)
        if Text1.value not in valores:
            valores.append(Text1.value)
            Dropdown1.options = valores
        if Dropdown4.value == 'Caso desconocido':          
            valores = list(Dropdown2.options)
            if Text2.value not in valores:
                valores.append(Text2.value)
                Dropdown2.options = valores
            valores = list(Dropdown5.options)
            if Text3.value not in valores:
                valores.append(Text3.value)
                Dropdown5.options = valores
        
def graficar(click):
    datos     = openpyxl.load_workbook(r'D:\Trabajo\Base_de_datos.xlsx')
    sheet_dat = datos.active
    valores   = [''] * (sheet_dat.max_row - 1)
    for i in range(2, sheet_dat.max_row + 1):
        cell_obj       = sheet_dat.cell(row = i, column = 2)
        mat            = cell_obj.value
        valores[i - 2] = str(mat)
    materias = list(set(valores))
    ocurrencias = [''] * len(materias)
    for i in range(len(materias)):
        ocurrencias[i] = valores.count(materias[i])
    bar.x = materias
    bar.y = ocurrencias
    ax_x.offset = {'scale':y_sc, 'value': max(ocurrencias)}

def acceder(value):
    if value == 'ozymandias':
        Password1.layout.display = 'None'
        HBox2.layout.visibility  = 'visible'

HTML1 = ipw.HTML(
        value  = f"<font size = '2'; font color = 'cyan'; font face = 'Courier New'><b><center>Nº de expediente:</center></b></font>",
        layout = ipw.Layout(width = '150px', height = '50px', padding = '0px', margin = '0px', border = '0px')
)   
HTML2 = ipw.HTML(
        value  = f"<font size = '2'; font color = 'cyan'; font face = 'Courier New'><b><center>Materia:</center></b></font>",
        layout = ipw.Layout(width = '150px', height = '50px', padding = '0px', margin = '0px', border = '0px')
)
HTML3 = ipw.HTML(
        value  = f"<font size = '2'; font color = 'cyan'; font face = 'Courier New'><b><center>Resolucion:</center></b></font>",
        layout = ipw.Layout(width = '150px', height = '50px', padding = '0px', margin = '0px', border = '0px')
)
HTML4 = ipw.HTML(
        value  = f"<font size = '2'; font color = 'cyan'; font face = 'Courier New'><b><center>URL:</center></b></font>",
        layout = ipw.Layout(width = '150px', height = '50px', padding = '0px', margin = '0px', border = '0px')
)
Text1 = ipw.Text(
        value       = '',
        description = '',
        layout      = ipw.Layout(width = '150px', height = '50px', padding = '0px', margin = '0px 50px 0px 50px', border = '0px'),
)
Text2 = ipw.Text(
        value       = '',
        description = '',
        layout      = ipw.Layout(width = '150px', height = '50px', padding = '0px', margin = '0px 50px 0px 50px', border = '0px'),
)
Text3 = ipw.Text(
        value       = '',
        description = '',
        layout      = ipw.Layout(width = '150px', height = '50px', padding = '0px', margin = '0px 50px 0px 50px', border = '0px'),
)
Text4 = ipw.Text(
        value       = '',
        description = '',
        layout      = ipw.Layout(width = '150px', height = '50px', padding = '0px', margin = '0px 50px 0px 50px', border = '0px'),
)

valores = [''] * (sheet_dat.max_row - 1)
for i in range(2, sheet_dat.max_row + 1):
    cell_obj       = sheet_dat.cell(row = i, column = 1)
    exp            = cell_obj.value
    valores[i - 2] = str(exp)
Dropdown1 = ipw.Dropdown(
            options     = valores,
            description = '',
            layout      = ipw.Layout(width = '150px', height = '30px', padding = '0px', margin = '0px 0px 20px 0px', border = '0px')
)
f_open_docx = ipw.interactive(open_docx, docx = Dropdown1)
Dropdown1.description = '';

valores = [''] * (sheet_dat.max_row - 1)
for i in range(2, sheet_dat.max_row + 1):
    cell_obj       = sheet_dat.cell(row = i, column = 2)
    mat            = cell_obj.value
    valores[i - 2] = str(mat)
Dropdown2 = ipw.Dropdown(
            options     = list(set(valores)),
            description = '',
            layout      = ipw.Layout(width = '150px', height = '30px', padding = '0px', margin = '0px 0px 20px 0px', border = '0px')
)
f_filtrar = ipw.interactive(filtrar, case = Dropdown2)
Dropdown2.description = '';

unico   = list(set(valores))[0]
indices = []
for valor in range(len(valores)):
    if valores[valor] == unico:
        indices.append(valor)
valores = [''] * (sheet_dat.max_row - 1)
for i in range(2, sheet_dat.max_row + 1):
    cell_obj       = sheet_dat.cell(row = i, column = 1)
    mat            = cell_obj.value
    valores[i - 2] = str(mat)
values  = [valores[i] for i in indices]
Dropdown3 = ipw.Dropdown(
            options     = values,
            description = '',
            layout      = ipw.Layout(width = '150px', height = '30px', padding = '0px', margin = '0px 0px 20px 0px', border = '0px')
)
f_open_docx = ipw.interactive(open_docx, docx = Dropdown3)
Dropdown3.description = '';

Dropdown4 = ipw.Dropdown(
            options     = ['Caso conocido', 'Caso desconocido'],
            description = '',
            layout      = ipw.Layout(width = '150px', height = '30px', padding = '0px', margin = '0px 0px 0px 50px', border = '0px')
)

valores = [''] * (sheet_dat.max_row - 1)
for i in range(2, sheet_dat.max_row + 1):
    cell_obj       = sheet_dat.cell(row = i, column = 3)
    res            = cell_obj.value
    valores[i - 2] = str(res)
Dropdown5 = ipw.Dropdown(
            options     = list(set(valores)),
            description = '',
            layout      = ipw.Layout(width = '150px', height = '30px', padding = '0px', margin = '0px 0px 20px 0px', border = '0px')
)

Button1 = ipw.Button(
          description  = 'Graficar',
          button_style = 'info',
          layout       = ipw.Layout(width = '150px', height = '30px', padding = '0px', margin = '0px', border = '0px')
)
Button1.on_click(graficar)

Button2 = ipw.Button(
          description  = 'Añadir',
          button_style = 'info',
          layout       = ipw.Layout(width = '150px', height = '30px', padding = '0px', margin = '0px', border = '0px')
)
Button2.on_click(add_data)

width  = "700px"
height = "700px"
urls   = "https://drive.google.com/file/d/1-k5dx76qY8V4-az8pxtYCNgQUqKAq_zT/preview"
HTML5  = ipw.HTML(
         value = f"<iframe src={urls} width={width} height={height}>",
)

valores = [''] * (sheet_dat.max_row - 1)
for i in range(2, sheet_dat.max_row + 1):
    cell_obj       = sheet_dat.cell(row = i, column = 2)
    mat            = cell_obj.value
    valores[i - 2] = str(mat)
materias = list(set(valores))
ocurrencias = [''] * len(materias)
for i in range(len(materias)):
    ocurrencias[i] = valores.count(materias[i])
x_ord   = bqp.OrdinalScale()
y_sc    = bqp.LinearScale()
bar     = bqp.Bars(x = materias, y = ocurrencias, scales = {'x': x_ord, 'y': y_sc},
          orientation = 'horizontal', colors = 'cyan', opacities = [0.5])
ax_x    = bqp.Axis(scale = x_ord, orientation = 'vertical', offset = {'scale':y_sc, 'value': max(ocurrencias)}, color = 'white', grid_color = 'black', grid_lines = 'dashed')
ax_y    = bqp.Axis(scale = y_sc, label = 'Numero de casos', label_color = 'red', tick_format = '0.0f', color = 'white', grid_color = 'cyan', grid_lines = 'dashed', label_offset = '-1em')
m_fig   = dict(left = 0, top = 30, bottom = 15, right = 0)
Figure1 = bqp.Figure(marks = [bar], axes = [ax_x, ax_y], padding_x = 0, padding_y = 0, animation_duration = 500, fig_margin = m_fig, layout = ipw.Layout(width='550px', height='475px'))
Figure1.background_style = {'fill': 'Black'}

Password1 = ipw.Password(
            value       = '',
            placeholder ='Ingrese la contraseña',
            description ='Contraseña:',
            layout      = ipw.Layout(width = '300px', height = '30px', padding = '0px', margin = '0px', border = '0px')
)
f_acceder = ipw.interactive(acceder, value = Password1)
display(Password1)

VBox1 = ipw.VBox([HTML1, HTML2, HTML3, HTML4, Button1],
        layout = ipw.Layout(width = 'auto', height = 'auto', align_items = 'flex-start')
)
VBox2 = ipw.VBox([Text1, Text2, Text3, Text4, Dropdown4],
        layout = ipw.Layout(width = 'auto', height = 'auto', align_items = 'flex-start')
)
VBox3 = ipw.VBox([Dropdown1, Dropdown2, Dropdown5, Dropdown3, Button2],
        layout = ipw.Layout(width = 'auto', height = 'auto', align_items = 'flex-start')
)
HBox1 = ipw.HBox([VBox1, VBox2, VBox3],
        layout = ipw.Layout(width = 'auto', height = 'auto', justify_content = 'flex-start')
)
VBox4 = ipw.VBox([HBox1, Figure1],
        layout = ipw.Layout(width = 'auto', height = 'auto', align_items = 'flex-start')
)
HBox2 = ipw.HBox([VBox4, HTML5],
        layout = ipw.Layout(width = 'auto', height = 'auto', justify_content = 'space-between')
)
    
display(HBox2)
HBox2.layout.visibility = "hidden"
```
